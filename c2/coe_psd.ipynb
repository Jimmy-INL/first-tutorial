```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_size_inches(12,6)


data=np.loadtxt('forceCoeffs.dat', skiprows=9)

f, Pxx_den = signal.periodogram(data[:,3], 1/0.05)

plt.xlim([0, 0.5])
plt.ylim([-1, 20])

plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')

plt.plot(f, Pxx_den)

#plt.semilogx()
#plt.semilogy()

plt.grid()
plt.show()

#fig.savefig('fre.png', dpi=300)
```
